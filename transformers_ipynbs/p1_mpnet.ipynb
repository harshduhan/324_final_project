```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances
import umap
```

```python
def load_manifesto(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()
    return text

def chunk_text(text, chunk_size=300):
    import re
    sentences = re.split(r'[.!?]\s+', text)
    chunks = []
    current_chunk = ""
    for sent in sentences:
        if len(current_chunk.split()) + len(sent.split()) < chunk_size:
            current_chunk += " " + sent
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sent
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks
```

```python
# Option 1: MiniLM
#model_name = 'all-MiniLM-L6-v2'
# Option 2: MPNet (better, slower)
model_name = 'all-mpnet-base-v2'

model = SentenceTransformer(model_name)
```

```python
data_dir = "masked_pdf_data"
manifestos = {}
embeddings = {}
labels = []

for party in ["BJP", "INC"]:
    for year in ["2009", "2014", "2019"]:
        file_path = os.path.join(data_dir, f"{party}_{year}.txt")
        print(f"Loading {file_path}")
        text = load_manifesto(file_path)
        chunks = chunk_text(text)
        print(f"{len(chunks)} chunks")

        embs = model.encode(chunks, batch_size=32, show_progress_bar=True)

        key = f"{party}_{year}"
        manifestos[key] = chunks
        embeddings[key] = embs
        labels.extend([key] * len(embs))

# Combine all embeddings
all_embeddings = np.vstack([emb for emb_list in embeddings.values() for emb in emb_list])
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
embeddings_2d = umap_model.fit_transform(all_embeddings)
all_labels = np.array(labels)
```

```python
# UMAP with fixed color palette

# Define custom palette
custom_palette = {
    'BJP_2009': "#F3CC4D",  # orange
    'BJP_2014': "#FF6A00",  # dark orange
    'BJP_2019': "#FF0000",  # reddish orange
    'INC_2009': "#00EAFF",  # steel blue
    'INC_2014': "#50A4F9",  # dodger blue
    'INC_2019': "#004EF7"   # deep sky blue
}

# Plot UMAP
plt.figure(figsize=(12,10))
sns.scatterplot(
    x=embeddings_2d[:,0], y=embeddings_2d[:,1], 
    hue=all_labels, palette=custom_palette, 
    s=80, alpha=0.7, edgecolor='black', linewidth=0.3
)

plt.title("Rhetorical Space of Indian Party Manifestos (UMAP of Sentence Embeddings)", fontsize=18)
plt.xlabel("UMAP Dimension 1", fontsize=14)
plt.ylabel("UMAP Dimension 2", fontsize=14)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=12, title="Party-Year")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("1MPNet_umap_scatter_masked.png", dpi=400)
plt.show()
```

```python
# Compute centroids in original embedding space
centroids = {}
for key, embs in embeddings.items():
    centroids[key] = np.mean(embs, axis=0)

# Compute drift distances
time_distances = {"BJP": [], "INC": []}

for party in ["BJP", "INC"]:
    years = ["2009", "2014", "2019"]
    for i in range(len(years)-1):
        k1 = f"{party}_{years[i]}"
        k2 = f"{party}_{years[i+1]}"
        d = cosine_distances([centroids[k1]], [centroids[k2]])[0][0]
        time_distances[party].append(d)

# Plot drift (BJP orange, INC blue)
party_line_colors = {"BJP": "#FF6A00", "INC": "#004EF7"}
plt.figure(figsize=(8,6))
for party in ["BJP", "INC"]:
    plt.plot(
        ["2009→2014", "2014→2019"],
        time_distances[party],
        marker="o",
        markersize=10,
        linewidth=2,
        label=party,
        color=party_line_colors[party],
    )

plt.xlabel("Time Period", fontsize=14)
plt.ylabel("Cosine Distance Between Manifesto Centroids", fontsize=14)
plt.title("Magnitude of Rhetorical Change Over Time", fontsize=18)
plt.legend(fontsize=12, title="Party")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("1MPNet_rhetorical_drift_masked.png", dpi=400)
plt.show()
# === Print actual delta values ===

print("\n=== Rhetorical Drift (Cosine Distance) ===")
for party in ["BJP", "INC"]:
    print(f"{party}:")
    for i, dist in enumerate(time_distances[party]):
        print(f"  {['2009→2014', '2014→2019'][i]}: {dist:.4f}")

# === Compare party total drift ===

total_drift = {}
for party in ["BJP", "INC"]:
    total_drift[party] = sum(time_distances[party])

print("\n=== Total Rhetorical Drift (2009→2019) ===")
for party in ["BJP", "INC"]:
    print(f"{party}: {total_drift[party]:.4f}")

# === Difference between parties ===

diff = total_drift["BJP"] - total_drift["INC"]
percent = (diff / total_drift["INC"]) * 100

print(f"\n=== BJP drift exceeds INC by {diff:.4f} ({percent:.1f}%) ===")

```

```python
# Logistic regression: train on 2009–2019, test on 2024 (raw text under `data/`).
# Cumulative accuracy / macro F1 over 2024 test paragraphs in order (BJP 2024, then INC 2024).

from pathlib import Path

from scipy.stats import t as student_t
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import cross_val_score

if "model" not in globals():
    raise RuntimeError("Run the cell that loads SentenceTransformer (`model = SentenceTransformer(...)`) first.")

data_root = Path("data")
train_years = ["2009", "2014", "2019"]
test_year = "2024"

manifestos = {}
embeddings = {}

train_blocks = []
y_train_list = []
for party in ["BJP", "INC"]:
    for year in train_years:
        path = data_root / f"{party}_{year}.txt"
        if not path.is_file():
            raise FileNotFoundError(f"Expected {path.resolve()}")
        text = load_manifesto(str(path))
        chunks = chunk_text(text)
        key = f"{party}_{year}"
        manifestos[key] = chunks
        embs = model.encode(chunks, batch_size=32, show_progress_bar=True)
        embeddings[key] = embs
        train_blocks.append(embs)
        y_train_list.extend([party] * len(chunks))

X_train = np.vstack(train_blocks)
y_train = np.array(y_train_list)


def cv_mean_ci(scores, alpha=0.05):
    """Two-sided CI for the mean CV score (t-distribution on k fold scores)."""
    scores = np.asarray(scores, dtype=float)
    n = len(scores)
    mean_s = float(scores.mean())
    sem = scores.std(ddof=1) / np.sqrt(n)
    t_crit = student_t.ppf(1 - alpha / 2, df=n - 1)
    margin = t_crit * sem
    return mean_s, mean_s - margin, mean_s + margin


clf_cv = LogisticRegression(max_iter=1000)
scores_bert = cross_val_score(clf_cv, X_train, y_train, cv=5, scoring="accuracy")
f1_bert = cross_val_score(clf_cv, X_train, y_train, cv=5, scoring="f1_macro")
acc_cv_mean, acc_cv_lo, acc_cv_hi = cv_mean_ci(scores_bert)
f1_cv_mean, f1_cv_lo, f1_cv_hi = cv_mean_ci(f1_bert)

test_blocks = []
y_test_ord_list = []
for party in ["BJP", "INC"]:
    path = data_root / f"{party}_{test_year}.txt"
    if not path.is_file():
        raise FileNotFoundError(f"Expected {path.resolve()}")
    text = load_manifesto(str(path))
    chunks = chunk_text(text)
    key = f"{party}_{test_year}"
    manifestos[key] = chunks
    embs = model.encode(chunks, batch_size=32, show_progress_bar=True)
    embeddings[key] = embs
    test_blocks.append(embs)
    y_test_ord_list.extend([party] * len(chunks))

X_test_ord = np.vstack(test_blocks)
y_test_ord = np.array(y_test_ord_list)

clf_bert = LogisticRegression(max_iter=1000)
clf_bert.fit(X_train, y_train)
y_pred_ord = clf_bert.predict(X_test_ord)

cum_acc = []
cum_f1 = []
for k in range(1, len(y_test_ord) + 1):
    cum_acc.append(accuracy_score(y_test_ord[:k], y_pred_ord[:k]))
    cum_f1.append(
        f1_score(y_test_ord[:k], y_pred_ord[:k], average="macro", zero_division=0)
    )

y_party = np.array(
    [key.split("_")[0] for key, chunk_list in manifestos.items() for _ in chunk_list]
)

print("=== MPNet + logistic regression: train 2009–2019, test 2024 (data/) ===")
print(f"Train paragraphs: {len(y_train)}, test paragraphs: {len(y_test_ord)}")
print(
    "5-fold CV on 2009–2019 (t-CI on fold means, same approach as TF-IDF): "
    f"accuracy {acc_cv_mean:.4f}  95% CI [{acc_cv_lo:.4f}, {acc_cv_hi:.4f}], "
    f"F1 macro {f1_cv_mean:.4f}  95% CI [{f1_cv_lo:.4f}, {f1_cv_hi:.4f}]"
)
print(
    f"2024 holdout (full train, metrics match final point on cumulative plot): "
    f"accuracy {cum_acc[-1]:.4f}, F1 macro {cum_f1[-1]:.4f}"
)

plt.figure(figsize=(10, 4))
plt.plot(range(1, len(cum_acc) + 1), cum_acc, label="Cumulative accuracy")
plt.plot(range(1, len(cum_f1) + 1), cum_f1, label="Cumulative F1 (macro)")
plt.xlabel("2024 test paragraph index (BJP then INC)")
plt.ylabel("Score")
plt.title("Streaming metrics on 2024 holdout (model trained on 2009–2019)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
